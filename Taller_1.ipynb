```python
import numpy as np
import matplotlib.pyplot as plt

def graficar(T, t_final, A, f, nombre_archivo='grafica_senoidal.png'):
    # Crear vector de tiempo
    t = np.arange(0, t_final+T, T)

    # Crear la señal senoidal
    y = A * np.sin(2 * np.pi * f * t)

    # Graficar
    plt.figure()
    plt.stem(t, y,)
    plt.plot(t,y,linewidth=1, color="r")
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.title('Señal Senoidal')
    plt.grid(True)
    plt.legend()

    # Guardar la gráfica como imagen
    plt.savefig(nombre_archivo)
    plt.show()
    plt.close()

    print(f'Imagen guardada como: {nombre_archivo}')

graficar(0.1, 10, 1, 0.5, "Figura 1")
#graficar(T=0.01, t_final=1, A=2, f=5)

# T:periodo de muestreo = f/muestras; f=numero de ciclos/tiempo total 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear figura
plt.figure(figsize=(10, 8))

# Lista de parámetros para ambas señales
parametros = [
    # [subplot_pos, T, Tf, A, f, linefmt, markerfmt, plot_color]
    [1, 0.05, 5, 1, 1, 'r', 'ro', 'g'],     # Señal 1: rojo/verde
    [2, 0.01, 0.5, 10, 5, 'g', 'go', 'r']   # Señal 2: verde/rojo
]

# Recorrer parámetros y graficar
for i, (pos, T, Tf, A, f, linefmt, markerfmt, plot_color) in enumerate(parametros):
    t = np.arange(0, Tf + T, T)
    y = A * np.sin(2 * np.pi * f * t)

    plt.subplot(2, 1, pos)
    plt.stem(t, y, linefmt=linefmt, markerfmt=markerfmt, basefmt='k')
    plt.plot(t, y, color=plot_color, linewidth=1.5)
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Amplitud")
    plt.grid(True)

    if pos == 1:
        plt.title("Comparación de dos senoidales")

# Ajustar y guardar
plt.tight_layout()
plt.savefig("figura_comparada_con_estilo.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_y_promediar(matriz):
    # Separar columnas
    tiempo = matriz[:, 0]
    amplitud = matriz[:, 1]

    # a. Graficar
    plt.plot(tiempo, amplitud, 'b-', linewidth=2)
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.title('Gráfica: Tiempo vs Amplitud')
    plt.grid(True)
    plt.show()

    # b. Calcular promedio de la amplitud
    promedio = np.mean(amplitud)

    return promedio
```

```python
X = np.array([0, 2, 5, 7, 15])
Y = np.array([3, 7, 10, 15, 26])

# Ajuste lineal con np.polyfit: el 1 indica que es una línea (grado 1)
pendiente, intercepto = np.polyfit(X, Y, 1)

# Generar la recta de ajuste
ajuste_lineal = pendiente * X + intercepto

# Graficar los puntos y la recta de ajuste
plt.scatter(X, Y, color='black', label='Datos')  # Nube de puntos
plt.plot(X, ajuste_lineal, 'r--', label='Ajuste lineal')  # Línea de ajuste

# Etiquetas y título
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ajuste Lineal de los Puntos')
plt.legend()

# Mostrar gráfica
plt.show()

# Imprimir pendiente e intercepto
print(f"Pendiente: {pendiente}")
print(f"Intercepto: {intercepto}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def curvaty(Tm, A, f, theta):
    # a. Definir el vector t
    t = np.arange(-4, 4, Tm)

    # b. Definir el vector y
    y = 5 * np.exp(-2 * t) + A * np.sin(2 * np.pi * f * t + theta)

    # c. Graficar
    plt.plot(t, y, 'r-', linewidth=2)  # Línea roja
    plt.grid(True)  # Rejilla
    plt.xlabel("Tiempo (t)")
    plt.ylabel("y")
    plt.title("Gráfica de t vs y")
    plt.show()

    return t, y

curvaty(0.1, 3, 1, 0)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Crear los dos vectores aleatorios
vector1 = np.random.uniform(1, 100, 1000)
vector2 = np.random.uniform(1, 100, 1000)

# a. Hallar el coeficiente de correlación entre los dos vectores
correlacion = np.corrcoef(vector1, vector2)
print("Coeficiente de correlación entre los vectores:")
print(correlacion)

# b. Hallar el ajuste lineal entre los dos vectores
# Usando np.polyfit para encontrar la pendiente (m) y el intercepto (b)
m, b = np.polyfit(vector1, vector2, 1)

# Generamos la línea de regresión
ajuste_lineal = m * vector1 + b

# c. Graficar los vectores 1 vs 2 con la línea de regresión
plt.scatter(vector1, vector2, label="Datos", color="black")
plt.plot(vector1, ajuste_lineal, label="Ajuste lineal", color="red", linestyle="--")
plt.xlabel("Vector 1 (independiente)")
plt.ylabel("Vector 2 (dependiente)")
plt.title("Ajuste Lineal entre Vector 1 y Vector 2")
plt.legend()
plt.grid(True)
plt.show()

# d. Calcular el error entre ambos vectores usando la expresión dada
N = len(vector1)  # Tamaño de los vectores
error = np.mean(vector1 - vector2)
print(f"El error entre los vectores es: {error}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
Tm = 0.05        # Periodo de muestreo
A = 10           # Amplitud
f = 1            # Frecuencia de la señal (1 Hz)

# Primer intervalo de tiempo: de 0 a 4.5 segundos
t1 = np.arange(0, 4.5 + Tm, Tm)
y1 = A * np.sin(2 * np.pi * f * t1)

# Segundo intervalo de tiempo: de 4.5 a 10 segundos
t2 = np.arange(4.5, 10 + Tm, Tm)
y2 = A * np.sin(2 * np.pi * f * t2)

# Graficar ambos intervalos en una sola figura
plt.plot(t1, y1, 'm--', linewidth=2)  # 'm--' es para línea morada discontinua
plt.plot(t2, y2, 'r--', linewidth=2)  # 'm--' es para línea morada discontinua

# Etiquetas y título
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("Gráfica de las Dos Funciones Senoidales")
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
