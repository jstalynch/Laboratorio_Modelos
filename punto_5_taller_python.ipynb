```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.figure(figsize=(10 , 10))
for i in range(1,7):
    theta = (i-1)*np.pi
    grosor = i/2
    A = 4
    Fs = 1
    Tm = 0.1
    Tf = 5
    t = np.arange(0, Tf+Tm, Tm)
    y = A*np.sin(2*np.pi*Fs*t+theta)
    plt.subplot(3,2,i)
    #plt.plot(t,y,)
    plt.stem(t,y)
    plt.plot(t,y,linewidth=grosor, color=np.random.rand(3))
    plt.title(f'theta={i-1}pi, grosor = {grosor}')
```
