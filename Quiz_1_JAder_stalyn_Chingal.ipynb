```python
import numpy as np
import matplotlib.pyplot as plt

t1 = np.arange(0,25+0.5,0.5) 
Y1 = 3 * np.sin(2 * np.pi * 0.1 * t1)

t2 = np.arange(0,5+0.1,0.1) 
Y2 = 2 * np.sin(2 * np.pi * 0.5 * t2)

matriz1 = np.array(np.vstack((t1,Y1)))
maz1t = matriz1.T
matriz2 = np.array(np.vstack((t2,Y2)))
maz2t = matriz2.T

def graficar(matrix1,matrix2):
    # Separar columnas
    tiempo1 = matrix1[:, 0]
    amplitud1 = matrix1[:, 1]

    tiempo2 = matrix2[:, 0]
    amplitud2 = matrix2[:, 1]


    plt.figure(figsize=(10, 8))
    plt.subplot(2, 1, 1)
    plt.stem(tiempo1, amplitud1, linefmt='r', markerfmt='ro', basefmt='k')  # palitos rojos, puntos rojos
    plt.plot(t1, amplitud1, color='g', linewidth=2)
    plt.title("cuerpo 1")
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid(True)
    

    plt.subplot(2, 1, 2)
    plt.stem(tiempo2, amplitud2, linefmt='g', markerfmt='go', basefmt='k')  # palitos verdes, puntos verdes
    plt.plot(t2, amplitud2, color='b', linewidth=2)                  # lnea roja continua
    plt.title("cuerpo 2")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Amplitud")
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    a = maz1t[:, 1]
    random_vector1 = np.random.randint(0, 21, size=len(a))

    rand_cuerpo1 = a + random_vector1

    b = maz2t[:, 1]
    rand_vector2 = np.random.randint(15, 30, size=len(b))

    rand_cuerpo2 = b + rand_vector2 

    # Calcular la regresin lineal (pendiente y coeficiente)
    m, b = np.polyfit(rand_cuerpo1, rand_cuerpo2, 1)

    # Graficar los puntos y la lnea de regresin
    plt.scatter(rand_cuerpo1, rand_cuerpo2, label='Datos')
    plt.plot(rand_cuerpo1, m*rand_cuerpo1 + b, color='red', label='Regresin lineal')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Regresin Lineal')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Imprimir los resultados de la regresin
    print(f"Pendiente: {m}")
    print(f"Intercepto: {b}")
    return m
   
#graficar(maz1t, maz2t)
m = graficar(maz1t, maz2t)


if m > 0:
    print(f"El movimiento de ambos cuerpos se relaciona: pendiente {m}")
else:
    print(f"El movimiento de ambos cuerpos no se relaciona: pendiente {m}")
```
